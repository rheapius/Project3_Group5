```python
import pandas as pd
```

```python
#read csv and create dataframe

file_path= "spotify-2023.csv"

try:
    df = pd.read_csv(file_path, encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding='latin1')

df.head(5)
```

```python
#combining the date columns

df['date_combined']= pd.to_datetime(df[['released_year','released_month','released_day']].rename(columns={'released_year': 'year', 'released_month': 'month', 'released_day': 'day'}))
df.head(5)
```

```python
# dropping columns

df= df.drop(columns=['artist_count','released_year','released_month','released_day'])
df.head(5)
```

```python
print(df.dtypes)
```

```python
#converting datatype to mumeric

convert=['streams','in_deezer_playlists','in_shazam_charts']
# Function to convert a value to numeric, ignoring errors
def convert_to_numeric(value):
    numeric_value=pd.to_numeric(value, errors='coerce')
    return numeric_value.fillna(0)

# Apply the function to specific columns
df[convert] = df[convert].apply(convert_to_numeric)

df['in_deezer_playlists'] = df['in_deezer_playlists'].astype(int)
df['in_shazam_charts'] = df['in_shazam_charts'].astype(int)
df['streams'] = df['streams'].astype(int)

print(df.dtypes)
```

```python
df.columns= ['Track','Artist','Spotify_playlists','Spotify_Charts','Streams','Apple_playlists','Apple_charts',
             'Deezer_playlists','Deezer_charts','Shazam_charts', 'BPM','Key','Mode',
             'Dancebility','Valence','Energy','Acousticness','Instrumentalness','Liveness',
             'Speechiness', 'Release_date']
```

```python
#creating dataframes

track_df= df[['Track', 'Artist', 'Release_date', 'Streams','BPM','Key','Mode',
             'Dancebility','Valence','Energy','Acousticness','Instrumentalness','Liveness',
             'Speechiness']]
spotify_df= df[['Track','Spotify_playlists','Spotify_Charts']]
apple_df= df[['Track','Apple_playlists','Apple_charts']]
deezer_df= df[['Track','Deezer_playlists','Deezer_charts']]
shazam_df= df[['Track', 'Shazam_charts']]
```

```python
#creating artist dataframe

artist_df= df.groupby('Artist').agg({'Streams':'sum','Dancebility': 'mean', 'Valence':'mean','Energy':'mean',
                                             'Acousticness':'mean','Instrumentalness':'mean','Liveness':'mean',
                                             'Speechiness':'mean'}).reset_index()

artist_df.columns= ['Artist','Streams','Dancebility','Valence','Energy','Acousticness','Instrumentalness','Liveness','Speechiness']

artist_df.head(5)
```

```python
#creating csv files

track_df.to_csv('track.csv', index=False)
artist_df.to_csv('artist.csv',index=False)
spotify_df.to_csv('spotify.csv', index=False)
apple_df.to_csv('apple.csv', index=False)
deezer_df.to_csv('deezer.csv', index=False)
shazam_df.to_csv('shazam.csv', index=False)
```
